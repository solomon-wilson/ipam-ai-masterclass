## RAG pattern

```python
knowledge_base = [
    "Leave applications must be approved by the line manager.",
    "Fuel usage must match approved route assignments.",
    "Audit files should include evidence, signatures, and dates."
]

def retrieve_context(question: str, knowledge: list[str]) -> str:
    """Return the most relevant text snippets using keyword overlap."""
    q_terms = set(question.lower().split())
    scored = sorted(
        knowledge,
        key=lambda item: len(q_terms.intersection(set(item.lower().split()))),
        reverse=True
    )
    return "\n".join(scored[:2])

question = "What should be in an audit file?"
context = retrieve_context(question, knowledge_base)
print("Context:\n", context)
```
