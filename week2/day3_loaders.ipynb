```python
"""
Utility functions used across notebooks.
These helpers are intentionally simple so students can read and modify them.
"""
import os
import json
from dataclasses import dataclass
from typing import Any, Dict, List, Optional

def pretty_json(data: Any) -> str:
    """Return a nicely formatted JSON string for learning and debugging."""
    return json.dumps(data, indent=2, ensure_ascii=False)

def safe_preview(text: str, n: int = 300) -> str:
    """Return the first n characters of a string for quick inspection."""
    return text[:n] + ("..." if len(text) > n else "")

print("Helper functions loaded.")
```

## Document processing pipeline

```python
sample_document = """
Sierra Leone Public Transport Authority
Compliance Notice

All operators must renew licenses before expiry.
Records should be kept for inspection.
Vehicle safety checks are mandatory every month.
"""

lines = [line.strip() for line in sample_document.splitlines() if line.strip()]
print("Loaded lines:", lines)
```

```python
# Simple chunking strategy for teaching purposes.
def chunk_text(text: str, chunk_size: int = 80) -> list[str]:
    """Split text into fixed-size chunks."""
    return [text[i:i + chunk_size] for i in range(0, len(text), chunk_size)]

chunks = chunk_text(sample_document, chunk_size=60)
for i, chunk in enumerate(chunks, start=1):
    print(f"Chunk {i}: {chunk!r}")
```

```python
# Example metadata structure that would accompany each chunk in a real loader.
chunk_metadata = {
    "source": "compliance_notice.txt",
    "department": "Operations",
    "document_type": "policy",
    "page": 1
}
print(pretty_json(chunk_metadata))
```
