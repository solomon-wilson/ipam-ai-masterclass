## Conceptual overview

```python
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity

# Example vectors used only for learning the similarity concept.
v1 = np.array([[0.1, 0.2, 0.7]])
v2 = np.array([[0.1, 0.25, 0.65]])
v3 = np.array([[0.9, 0.1, 0.0]])

print("Similarity v1-v2:", cosine_similarity(v1, v2)[0][0])
print("Similarity v1-v3:", cosine_similarity(v1, v3)[0][0])
```

```python
# A simple mental model:
# - similar meanings should end up close together in vector space
# - unrelated meanings should be farther apart
texts = [
    "bus route delay",
    "vehicle maintenance issue",
    "student registration form"
]
print("Sample texts prepared for embedding experiments:")
for t in texts:
    print("-", t)
```
