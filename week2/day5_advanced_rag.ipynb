## Advanced retrieval ideas

```python
routes = {
    "hr": ["leave policy", "staff conduct", "recruitment procedure"],
    "finance": ["budget approval", "payment voucher", "procurement"],
    "it": ["password reset", "network issue", "system access"]
}

def route_question(question: str) -> str:
    """Pick the best domain based on keyword overlap."""
    q = question.lower()
    best_domain = None
    best_score = -1
    for domain, keywords in routes.items():
        score = sum(1 for k in keywords if k in q)
        if score > best_score:
            best_domain = domain
            best_score = score
    return best_domain or "general"

question = "How do I request a password reset?"
print("Route:", route_question(question))
```
