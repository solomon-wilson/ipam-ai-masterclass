## Local vector store simulation

```python
# This notebook uses in-memory examples so it can run even in constrained environments.
documents = [
    {"id": "doc1", "text": "All staff must submit leave forms before travel.", "source": "HR Policy"},
    {"id": "doc2", "text": "Vehicles must undergo weekly maintenance checks.", "source": "Fleet Policy"},
    {"id": "doc3", "text": "All audit evidence must be retained for seven years.", "source": "Audit Manual"},
]

query = "How long should audit evidence be kept?"
print("Query:", query)
print("Candidate documents:")
for doc in documents:
    print("-", doc["source"], "=>", doc["text"])
```

```python
# In a full implementation, these documents would be embedded and stored in ChromaDB.
# Here we show the retrieval pattern with simple Python logic.
def keyword_score(query: str, text: str) -> int:
    query_terms = set(query.lower().split())
    text_terms = set(text.lower().split())
    return len(query_terms.intersection(text_terms))

ranked = sorted(
    documents,
    key=lambda d: keyword_score(query, d["text"]),
    reverse=True
)

print("Best match:", ranked[0]["source"])
print(ranked[0]["text"])
```
