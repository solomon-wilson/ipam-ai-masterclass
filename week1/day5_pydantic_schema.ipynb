```python
"""
Utility functions used across notebooks.
These helpers are intentionally simple so students can read and modify them.
"""
import os
import json
from dataclasses import dataclass
from typing import Any, Dict, List, Optional

def pretty_json(data: Any) -> str:
    """Return a nicely formatted JSON string for learning and debugging."""
    return json.dumps(data, indent=2, ensure_ascii=False)

def safe_preview(text: str, n: int = 300) -> str:
    """Return the first n characters of a string for quick inspection."""
    return text[:n] + ("..." if len(text) > n else "")

print("Helper functions loaded.")
```

## Structured schemas

```python
from pydantic import BaseModel, Field
from typing import List

class IncidentReport(BaseModel):
    """Schema for a simple incident report extracted from text."""
    incident_type: str = Field(description="Type of incident")
    department: str = Field(description="Responsible department")
    severity: str = Field(description="Low, medium, or high")
    actions: List[str] = Field(description="Recommended actions")

sample = IncidentReport(
    incident_type="Network outage",
    department="IT",
    severity="High",
    actions=["Restart network equipment", "Escalate to ISP", "Document downtime"]
)
print(sample.model_dump_json(indent=2))
```

```python
# Simulated extraction result that can later be produced by an LLM.
raw_data = {
    "incident_type": "Late dispatch",
    "department": "Operations",
    "severity": "medium",
    "actions": [
        "Review dispatcher logs",
        "Notify route manager",
        "Track recurrence"
    ]
}

report = IncidentReport(**raw_data)
print(report)
```
