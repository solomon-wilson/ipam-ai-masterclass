## Prompt design patterns

```python
# A simple prompt template for consistent output.
template = """
You are an assistant helping a transport authority analyst.

Task:
{task}

Output format:
1. Summary
2. Risks
3. Recommended action
"""

prompt = template.format(
    task="Review a late bus dispatch and highlight possible operational risks."
)
print(prompt)
```
