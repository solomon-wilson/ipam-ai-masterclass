```python
"""
Utility functions used across notebooks.
These helpers are intentionally simple so students can read and modify them.
"""
import os
import json
from dataclasses import dataclass
from typing import Any, Dict, List, Optional

def pretty_json(data: Any) -> str:
    """Return a nicely formatted JSON string for learning and debugging."""
    return json.dumps(data, indent=2, ensure_ascii=False)

def safe_preview(text: str, n: int = 300) -> str:
    """Return the first n characters of a string for quick inspection."""
    return text[:n] + ("..." if len(text) > n else "")

print("Helper functions loaded.")
```

## Prompt templates

```python
from langchain_core.prompts import ChatPromptTemplate

prompt = ChatPromptTemplate.from_messages([
    ("system", "You are a careful assistant that writes concise enterprise summaries."),
    ("human", "Summarize this report in three sentences: {report}")
])

formatted = prompt.format_messages(
    report="The generator was offline for 2 hours due to scheduled maintenance."
)
for message in formatted:
    print(message.type, "=>", message.content)
```

## Output parsing

```python
from langchain_core.output_parsers import StrOutputParser

parser = StrOutputParser()
chain = prompt | llm | parser

result = chain.invoke({
    "report": "The office printer failed during records printing."
})
print(result)
```
