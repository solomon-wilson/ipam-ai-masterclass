```python
"""
Utility functions used across notebooks.
These helpers are intentionally simple so students can read and modify them.
"""
import os
import json
from dataclasses import dataclass
from typing import Any, Dict, List, Optional

def pretty_json(data: Any) -> str:
    """Return a nicely formatted JSON string for learning and debugging."""
    return json.dumps(data, indent=2, ensure_ascii=False)

def safe_preview(text: str, n: int = 300) -> str:
    """Return the first n characters of a string for quick inspection."""
    return text[:n] + ("..." if len(text) > n else "")

print("Helper functions loaded.")
```

## Course context

In this course, you will build practical AI systems that support enterprise work:
compliance lookup, policy assistants, audit support, and autonomous agents.

```python
# Securely load the Gemini key in Colab.
# If you are not in Colab, set GOOGLE_API_KEY in your environment manually.
import os

try:
    from google.colab import userdata
    os.environ["GOOGLE_API_KEY"] = userdata.get("GEMINI_API_KEY")
    print("Gemini API key loaded from Colab Secrets.")
except Exception as exc:
    print("Colab secrets not available. Set GOOGLE_API_KEY manually.")
    print(f"Details: {exc}")

try:
    import google.generativeai as genai
    from langchain_google_genai import ChatGoogleGenerativeAI
    if os.getenv("GOOGLE_API_KEY"):
        genai.configure(api_key=os.environ["GOOGLE_API_KEY"])
        llm = ChatGoogleGenerativeAI(model="gemini-1.5-flash", temperature=0)
        response = llm.invoke("What is the capital of Sierra Leone?")
        print(response.content)
    else:
        print("GOOGLE_API_KEY is not set yet.")
except Exception as exc:
    print("Model test not run successfully.")
    print(f"Details: {exc}")
```
