## Agent design

```python
class SimpleAgent:
    """A minimal educational agent that picks one action at a time."""

    def __init__(self):
        self.history = []

    def run(self, question: str) -> str:
        """Decide how to handle the question and record it in history."""
        self.history.append(question)
        q = question.lower()
        if "route" in q:
            return "Action: query_routes_database"
        if "leave" in q:
            return "Action: check_hr_policy"
        return "Action: answer_directly"

agent = SimpleAgent()
for q in [
    "What is the route status?",
    "How do I apply for leave?",
    "Explain what LangChain does."
]:
    print(q, "=>", agent.run(q))
print("History:", agent.history)
```
