```python
"""
Utility functions used across notebooks.
These helpers are intentionally simple so students can read and modify them.
"""
import os
import json
from dataclasses import dataclass
from typing import Any, Dict, List, Optional

def pretty_json(data: Any) -> str:
    """Return a nicely formatted JSON string for learning and debugging."""
    return json.dumps(data, indent=2, ensure_ascii=False)

def safe_preview(text: str, n: int = 300) -> str:
    """Return the first n characters of a string for quick inspection."""
    return text[:n] + ("..." if len(text) > n else "")

print("Helper functions loaded.")
```

## Tool examples

```python
def get_route_status(route_number: str) -> dict:
    """Simulate a live route status tool."""
    return {
        "route_number": route_number,
        "status": "On time",
        "last_update": "10:45 AM"
    }

def calculate_delay(minutes_late: int) -> str:
    """Convert a delay into a human-readable message."""
    if minutes_late <= 5:
        return "Minor delay"
    if minutes_late <= 20:
        return "Moderate delay"
    return "Severe delay"

print(get_route_status("R12"))
print(calculate_delay(17))
```

```python
# Tool selection can be implemented using simple if/else logic.
def handle_user_request(text: str):
    """Dispatch to the right tool based on the user's request."""
    text_lower = text.lower()
    if "route" in text_lower:
        return get_route_status("R12")
    if "delay" in text_lower:
        return calculate_delay(17)
    return "No tool needed."

print(handle_user_request("Check the route status"))
```
