## Conversation buffer

```python
from collections import deque

class ConversationBuffer:
    """Store the last N turns of a conversation for context injection."""

    def __init__(self, max_turns: int = 5):
        # Limit history to avoid exceeding the model context window.
        self._turns = deque(maxlen=max_turns)

    def add(self, role: str, content: str) -> None:
        """Append one turn to the buffer."""
        self._turns.append({"role": role, "content": content})

    def as_text(self) -> str:
        """Return the buffer as a formatted string for prompt injection."""
        return "\n".join(
            f"{t['role'].capitalize()}: {t['content']}"
            for t in self._turns
        )

buf = ConversationBuffer(max_turns=3)
buf.add("user", "What is the fuel policy?")
buf.add("assistant", "Fuel must match approved route assignments.")
buf.add("user", "Who approves the routes?")
print(buf.as_text())
```

```python
def answer_with_memory(question: str, buffer: ConversationBuffer) -> str:
    """Build a prompt that includes conversation history, then query the model."""
    history_text = buffer.as_text()
    prompt = f"""
    You are a helpful transport authority assistant.

    Conversation so far:
    {history_text}

    New question: {question}
    """
    # In a live notebook, replace this placeholder with ask_llm(prompt).
    print("Prompt sent to model:")
    print(prompt)
    return "(model response placeholder)"

buf = ConversationBuffer()
buf.add("user", "What documents are needed for a vehicle inspection?")
buf.add("assistant", "You need the log book, insurance, and roadworthiness certificate.")
answer_with_memory("How long are those certificates valid?", buf)
```
