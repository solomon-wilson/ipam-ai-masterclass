```python
"""
Utility functions used across notebooks.
These helpers are intentionally simple so students can read and modify them.
"""
import os
import json
from dataclasses import dataclass
from typing import Any, Dict, List, Optional

def pretty_json(data: Any) -> str:
    """Return a nicely formatted JSON string for learning and debugging."""
    return json.dumps(data, indent=2, ensure_ascii=False)

def safe_preview(text: str, n: int = 300) -> str:
    """Return the first n characters of a string for quick inspection."""
    return text[:n] + ("..." if len(text) > n else "")

print("Helper functions loaded.")
```

## ReAct concept

```python
# A minimal ReAct-style example using plain Python.
# Reasoning: decide what information is missing.
# Acting: call a tool or fetch data.
def decide_action(question: str) -> str:
    if "weather" in question.lower():
        return "call_weather_api"
    if "route" in question.lower():
        return "query_routes_database"
    return "answer_directly"

for q in [
    "What is the weather in Freetown?",
    "Show the current route status.",
    "Explain what an AI agent is."
]:
    print(q, "=>", decide_action(q))
```

```python
# ReAct is useful when the model needs to decide between thinking and using tools.
reasoning_trace = [
    "Identify the user's intent.",
    "Check whether external data is required.",
    "Select the appropriate tool or response."
]
print("\n".join(f"- {step}" for step in reasoning_trace))
```
